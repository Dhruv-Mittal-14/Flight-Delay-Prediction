```python
#Import Libraries
import pandas as pd
import numpy as np
import joblib
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler

# Data Splitting
from sklearn.model_selection import train_test_split

# Encoding
from sklearn.preprocessing import LabelEncoder

# Models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

# Evaluation
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)
```

```python
#Load Data
df = pd.read_csv("D:/Flight Delay Prediction/data/processed/flight_features.csv")

print("Dataset Shape:", df.shape)

df.head()
```

```python
#Feature Selection
drop_columns = [
    "flight_id",
    "flight_number",
    "flight_date",
    "arr_delay",
    "dep_delay",
    "month",
    "quarter",
    "flight_duration_hours",
    "delay_severity",
    "created_at",
    "crs_dep_time"
]

df = df.drop(columns=drop_columns)

df.head()
```

```python
#Feature and Target
X = df.drop(columns=["delay_status"])

y = df["delay_status"]

print("Features Shape :", X.shape)
print("Target Shape :", y.shape)
```

```python
categorical_features = [
    "marketing_airline_network",
    "origin",
    "dest",
    "day_name",
    "departure_period",
    "distance_category",
    "duration_category"
]

numerical_features = [
    "distance",
    "crs_elapsed_time",
    "day",
    "day_of_week",
    "is_weekend",
    "departure_hour",
    "departure_minute",
    "flight_duration_minutes"
]
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        (
            "num",
            StandardScaler(),
            numerical_features
        ),

        (
            "cat",
            OneHotEncoder(handle_unknown="ignore"),
            categorical_features
        )
    ]
)
```

```python
#Train-Test Data
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)


print("Training Size :", X_train.shape)
print("Testing Size :", X_test.shape)
```

```python
#Class Distribution
print(y_train.value_counts(normalize=True) * 100)
```

```python
#Columns used for training
print("Features Used for Training:")
print(X.columns.tolist())
```

```python
# Logistic Regression Model

logistic_pipeline = Pipeline(
    steps=[
        ("preprocessor", preprocessor),

        (
            "classifier",

            LogisticRegression(
                class_weight="balanced",
                random_state=42,
                max_iter=3000
            )
        )
    ]
)
```

```python
#Train
logistic_pipeline.fit(
    X_train,
    y_train
)
```

```python
#Predict
y_pred_lr = logistic_pipeline.predict(X_test)

#Probability
y_prob_lr = logistic_pipeline.predict_proba(X_test)

lr_accuracy = accuracy_score(y_test, y_pred_lr)
lr_precision = precision_score(y_test, y_pred_lr)
lr_recall = recall_score(y_test, y_pred_lr)
lr_f1 = f1_score(y_test, y_pred_lr)

print(f"Accuracy : {lr_accuracy:.4f}")
print(f"Precision: {lr_precision:.4f}")
print(f"Recall   : {lr_recall:.4f}")
print(f"F1 Score : {lr_f1:.4f}")
```

```python
print(confusion_matrix(y_test, y_pred_lr))
```

```python
print(classification_report(y_test, y_pred_lr))
```

```python
lr_results = {
    "Model":"Logistic Regression",
    "Accuracy":lr_accuracy,
    "Precision":lr_precision,
    "Recall":lr_recall,
    "F1 Score":lr_f1
}
```

```python
#Decision Tree
decision_pipeline = Pipeline(
    steps=[
        ("preprocessor", preprocessor),

        (
            "classifier",

            DecisionTreeClassifier(
                class_weight="balanced",
                random_state=42,
                max_depth=15
            )
        )
    ]
)
```

```python
#Train
decision_pipeline.fit(
    X_train,
    y_train
)
```

```python
#Predict
y_pred_dt = decision_pipeline.predict(X_test)

#Probability
y_prob_dt = decision_pipeline.predict_proba(X_test)

dt_accuracy = accuracy_score(y_test, y_pred_dt)
dt_precision = precision_score(y_test, y_pred_dt)
dt_recall = recall_score(y_test, y_pred_dt)
dt_f1 = f1_score(y_test, y_pred_dt)

print(f"Accuracy : {dt_accuracy:.4f}")
print(f"Precision: {dt_precision:.4f}")
print(f"Recall   : {dt_recall:.4f}")
print(f"F1 Score : {dt_f1:.4f}")
```

```python
print(confusion_matrix(y_test, y_pred_dt))
```

```python
print(classification_report(y_test, y_pred_dt))
```

```python
dt_results = {
    "Model":"Decision Tree",
    "Accuracy":dt_accuracy,
    "Precision":dt_precision,
    "Recall":dt_recall,
    "F1 Score":dt_f1
}
```

```python
#Random Forest
random_pipeline = Pipeline(
    steps=[
        ("preprocessor", preprocessor),

        ("classifier",
         RandomForestClassifier(
             n_estimators=300,
             max_depth=20,
             min_samples_split=5,
             min_samples_leaf=2,
             class_weight="balanced_subsample",
             random_state=42,
             n_jobs=-1
         ))
    ]
)
```

```python
#Train
random_pipeline.fit(
    X_train,
    y_train
)
```

```python
#Predict
y_pred_rf = random_pipeline.predict(X_test)
# print(pd.Series(y_pred_rf).value_counts())
# print(y_test.value_counts())
#Probability
y_prob_rf = random_pipeline.predict_proba(X_test)
# print(confusion_matrix(y_test, y_pred_rf))

rf_accuracy = accuracy_score(y_test, y_pred_rf)
rf_precision = precision_score(y_test, y_pred_rf)
rf_recall = recall_score(y_test, y_pred_rf)
rf_f1 = f1_score(y_test, y_pred_rf)

print(f"Accuracy : {rf_accuracy:.4f}")
print(f"Precision: {rf_precision:.4f}")
print(f"Recall   : {rf_recall:.4f}")
print(f"F1 Score : {rf_f1:.4f}")
```

```python
print(confusion_matrix(y_test, y_pred_rf))
```

```python
print(classification_report(y_test, y_pred_rf))
```

```python
rf_results = {
    "Model":"Random Forest",
    "Accuracy":rf_accuracy,
    "Precision":rf_precision,
    "Recall":rf_recall,
    "F1 Score":rf_f1
}
```

```python
#Gradient Boosting
gradient_pipeline = Pipeline(
    steps=[
        ("preprocessor", preprocessor),

        (
            "classifier",

            GradientBoostingClassifier(
                random_state=42
            )
        )
    ]
)
```

```python
#Train
gradient_pipeline.fit(
    X_train,
    y_train
)
```

```python
#Predict
y_pred_gb = gradient_pipeline.predict(X_test)

#Probability
y_prob_gb = gradient_pipeline.predict_proba(X_test)

gb_accuracy = accuracy_score(y_test, y_pred_gb)
gb_precision = precision_score(y_test, y_pred_gb)
gb_recall = recall_score(y_test, y_pred_gb)
gb_f1 = f1_score(y_test, y_pred_gb)

print(f"Accuracy : {gb_accuracy:.4f}")
print(f"Precision: {gb_precision:.4f}")
print(f"Recall   : {gb_recall:.4f}")
print(f"F1 Score : {gb_f1:.4f}")
```

```python
print(confusion_matrix(y_test, y_pred_gb))
```

```python
print(classification_report(y_test, y_pred_gb))
```

```python
gb_results = {
    "Model":"Gradient Boosting",
    "Accuracy":gb_accuracy,
    "Precision":gb_precision,
    "Recall":gb_recall,
    "F1 Score":gb_f1
}
```
